**STO EPOMENO CELL TREXW MAZIKA GIA !ENA PROION! POU EXW HDH EPILEKSEI PROHGOUMENWS KAI EXEI GINEI BALANCE TO TRAIN DATASET, OLES TIS DIAFORETIKES PARAMETROUS. TA paotelesmata ta grafw se ena pandas df to opoiw to apothikeuw sto telos**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, roc_auc_score, accuracy_score

import warnings

# Suppress the deprecation warning for pd.append()
warnings.filterwarnings('ignore', message='The frame.append method') 
```

```python
#params for random forest


criterio=['gini','entropy']
n_estimator=[100, 250, 500, 1000]
max_feature=[0.6, 0.75, 0.9, 1.0]


results = pd.DataFrame(columns=['criterion', 'n_estimators', 'max_features', 'AUC', 'accuracy', 'precision', 'recall'])
for kritirio in criterio:
    for n_estimato in n_estimator:
        for max_featur in max_feature:
            #train the model
            # create a RandomForest model
            
            model = RandomForestClassifier( criterion=kritirio, n_estimators=n_estimato, max_features=max_featur)
            # fit the model to the data
            model.fit(features_train, target_train)
            #==========================================================================================================================================
            # make predictions on new data
            # predict class labels for the testing set
            predictions = model.predict(features_test)
            y_pred_prob = model.predict_proba(features_test)
            # print the predicted values
            '''print('The predictions are:')
            print(predictions)'''

            #==========================================================================================================================================
            # compute AUC score
            auc = roc_auc_score(target_test, y_pred_prob[:, 1])
            #print("ROC AUC on test data: {:.4f}".format(auc))

            # compute accuracy score
            accuracy = accuracy_score(target_test, predictions)
            #print("Accuracy: {:.4f}".format(accuracy))

            # compute precision score
            precision = precision_score(target_test, predictions)
            #print("Precision: {:.4f}".format(precision))

            # compute recall score
            recall = recall_score(target_test, predictions)
            #print("Recall: {:.4f}".format(recall))
            results=results.append({'criterion': kritirio, 'n_estimators': n_estimato, 'max_features': max_featur, 'AUC':auc, 'accuracy':accuracy, 'precision':precision, 'recall':recall}, ignore_index=True)
file_path='C:/Users/Y9FHP3726/Desktop/ANDREAS/DiplomaThesis/experiments/RandomForestResults/27_04_2023_results_Random_Forest_'+product+'.csv'
results.to_csv(file_path, index=False)
```
